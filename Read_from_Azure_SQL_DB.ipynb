**In order to read the data from database stored on Azure we have to create a connection string which require different parameter values <br/>
We get those values from our Azure on whcih we created our SQL Database along with Tables**<br/>
## Method 1
**Using JDBC connection using parameters**

```
jdbcHostname = "ss-owaisde.database.windows.net"     # This is our server name on Azure SQL
jdbcPort = 1433    # This is common port 
jdbcDatabase = "database-owaisde"     # This is our database name on Azure SQL
jdbcUsername = "owaisde"              # Username we have given while creating server on Azure SQL
jdbcPassword = "test@1234"            # Password we have given while creating server on Azure SQL
jdbcDriver = "com.microsoft.sqlserver.jdbc.SQLServerDriver"      # This parameter is constant others we get from Azure

jdbcUrl = f"jdbc:sqlserver://{jdbcHostname}:{jdbcPort};databaseName={jdbcDatabase};user{jdbcUsername};password={jdbcPassword};driver={jdbcDriver}"
```

```
df1 = spark.read.format("jdbc").option("url",jdbcUrl).option("dbtable", "SalesLT.Product").load()    
df1.head()
```

## Method 2
**Using connection string directly which we can get from Azure SQL Database**

```
connectionString = "jdbc:sqlserver://ss-owaisde.database.windows.net:1433;database=database-owaisde;user=owaisde@ss-owaisde;password={test@1234};encrypt=true;trustServerCertificate=false;hostNameInCertificate=*.database.windows.net;loginTimeout=30;"
```

```
df = spark.read.jdbc(connectionString, "SalesLT.Product")
df.head()
```

In order to execute the above command we need to add the databricks IP Address on Azure resource. After that we can run the command again as shown below

```
df = spark.read.jdbc(connectionString, "SalesLT.Product")
df.show(5)   # displaying first 5 rows from the dataframe
```

```
df = spark.read.jdbc(connectionString, "SalesLT.Address")  # Data from another table called Address
df.show(5)   # displaying first 5 rows from the dataframe
```
